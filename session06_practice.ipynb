```python
import string
# test_str = 'i need to remove punctuatuion! ;'
test_str = rabbit_text.lower()
clean_str = test_str.translate (str.maketrans('', '', string.punctuation))
print(clean_str) 
#removing punc
```

```python
split_str = clean_str.split()
```

```python
len(split_str)
```

```python
words_set = set(split_str)
```

```python
words_set
```

```python
len(words_set)
```

```python
word_count_dict = {}
for word in words_set:
    # word_str = str(word)
    word_count = clean_str.count(word)
    word_count_dict[word] = word_count
```
